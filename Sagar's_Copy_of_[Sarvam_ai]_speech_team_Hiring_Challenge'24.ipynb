````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import os
import requests
import time

# Create a directory to store the downloaded files
if not os.path.exists("new_testament"):
    os.makedirs("new_testament")

# Initialize the base URL
base_url = "https://live.bible.is/bible/HINOHC/MAT/1"  # Starting from Matthew

# Set up the Selenium webdriver
# driver = webdriver.Chrome(executable_path='/path/to/chromedriver')  # Update the path accordingly

def fetch_content(driver, chapter_folder, chapter):
    # Wait for the text and audio elements to be present
    wait = WebDriverWait(driver, 20)  # Wait up to 20 seconds

    # Wait for the audio element
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'video.audio-player')))

    # Wait for the text element
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div.chapter.justify')))

    # Get the HTML content
    html = driver.page_source

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html, "html.parser")

    # Find the audio file URL using BeautifulSoup
    audio_element = soup.find("video", class_='audio-player')
    audio_url = audio_element['src'] if audio_element else None

    # Find the textual content using BeautifulSoup
    text_element = soup.find("div", class_='chapter justify')
    text_content = text_element.text if text_element else None

    if audio_url:
        # Download the audio file
        audio_response = requests.get(audio_url)
        with open(f"{chapter_folder}/chapter_{chapter}.mp3", "wb") as f:
            f.write(audio_response.content)
    else:
        print(f"No audio file downloaded for chapter {chapter}!")

    if text_content:
        # Save the textual content to a file
        with open(f"{chapter_folder}/chapter_{chapter}.txt", "w", encoding="utf-8") as f:
            f.write(text_content)
    else:
        print(f"No textual content saved for chapter {chapter}!")

    print(f"Chapter {chapter} processed successfully!\n")

try:
    chapter = 1
    current_url = base_url
    while chapter <= 260:
        # Create a folder for the current chapter
        chapter_folder = f"new_testament/chapter_{chapter}"
        if not os.path.exists(chapter_folder):
            os.makedirs(chapter_folder)

        # Load the page using Selenium
        driver.get(current_url)

        retries = 3
        for attempt in range(retries):
            try:
                fetch_content(driver, chapter_folder, chapter)
                break
            except Exception as e:
                print(f"Attempt {attempt + 1} failed: {e}")
                if attempt < retries - 1:
                    print("Retrying...")
                    time.sleep(5)  # Wait before retrying
                    driver.get(current_url)  # Reload the page
                else:
                    print("Max retries reached. Moving to the next chapter.")

        # Attempt to find the "Next" button and click it using By.XPATH
        try:
            wait = WebDriverWait(driver, 20)  # Wait up to 20 seconds
            next_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//div[@title='Next Chapter' and @id='next-chapter-audio']")))
            next_button.click()
            current_url = driver.current_url  # Update the current URL to the next chapter's URL
        except Exception as e:
            print(f"Next button not found or an error occurred: {e}. Moving to the next iteration.")

        chapter += 1
except Exception as e:
    print(f"An error occurred: {e}")
finally:
    # Close the Selenium webdriver
    driver.quit()
````

````python
import subprocess

def convert_mp3_to_wav(mp3_file_path, wav_file_path):
    command = ['ffmpeg', '-i', mp3_file_path, wav_file_path]
    subprocess.call(command)

mp3_file_path = '/content/new_testament/chapter_3/chapter_3.mp3'
wav_file_path = 'converted_audio.wav'

convert_mp3_to_wav(mp3_file_path, wav_file_path)
````

````python
import os
import numpy as np
from scipy.io import wavfile
from numpy import log10

def signaltonoiseScipy(a, axis=0, ddof=0):
    a = np.asanyarray(a)
    m = a.mean(axis)
    sd = a.std(axis=axis, ddof=ddof)
    return np.where(sd == 0, 0, m / sd)

def snr(file):
    if os.path.isfile(file):
        rate, data = wavfile.read(file)

        # Check if the audio is stereo and convert to mono if needed
        if len(data.shape) > 1:
            data = np.mean(data, axis=1)

        # Normalize the data
        norm = data / np.max(np.abs(data))

        # Calculate SNR using signaltonoiseScipy function
        snr_value = signaltonoiseScipy(norm)

        # Convert SNR to decibels (dB)
        snr_db = 20 * log10(abs(snr_value))

        return snr_db

# Example usage
file_path = '/content/converted_audio.wav'
snr_value = snr(file_path)
print("SNR: {} dB".format(snr_value))
````

````python
import os
import numpy as np
import scipy.signal as sig
import scipy.linalg as linalg
import matplotlib.pyplot as plt
from scipy.io import wavfile
from numpy import log10, argmax, array, argmin, mod, floor, median, copy

def bandpower(ps, mode='psd'):
    """
    Estimate bandpower, see https://de.mathworks.com/help/signal/ref/bandpower.html
    """
    if mode == 'time':
        x = ps
        l2norm = linalg.norm(x) ** 2 / len(x)
        return l2norm
    elif mode == 'psd':
        return sum(ps)

def getIndizesAroundPeak(arr, peakIndex, searchWidth=1000):
    peakBins = []
    magMax = arr[peakIndex]
    curVal = magMax
    for i in range(searchWidth):
        newBin = peakIndex + i
        newVal = arr[newBin]
        if newVal > curVal:
            break
        else:
            peakBins.append(int(newBin))
            curVal = newVal
    curVal = magMax
    for i in range(searchWidth):
        newBin = peakIndex - i
        newVal = arr[newBin]
        if newVal > curVal:
            break
        else:
            peakBins.append(int(newBin))
            curVal = newVal
    return array(list(set(peakBins)))

def freqToBin(fAxis, Freq):
    return argmin(abs(fAxis - Freq))

def getPeakInArea(psd, faxis, estimation, searchWidthHz=10):
    """
    Returns bin and frequency of the maximum in an area
    """
    binLow = freqToBin(faxis, estimation - searchWidthHz)
    binHi = freqToBin(faxis, estimation + searchWidthHz)
    peakbin = binLow + argmax(psd[binLow:binHi])
    return peakbin, faxis[peakbin]

def getHarmonics(fundFrequency, sr, nHarmonics=6, aliased=False):
    harmonicMultipliers = np.arange(2, nHarmonics + 2)
    harmonicFs = fundFrequency * harmonicMultipliers
    if not aliased:
        harmonicFs[harmonicFs > sr / 2] = -1
        harmonicFs = np.delete(harmonicFs, harmonicFs == -1)
    else:
        nyqZone = floor(harmonicFs / (sr / 2))
        oddEvenNyq = nyqZone % 2
        harmonicFs = mod(harmonicFs, sr / 2)
        harmonicFs[oddEvenNyq == 1] = (sr / 2) - harmonicFs[oddEvenNyq == 1]
    return harmonicFs

def plotSNR(ps, faxis, fundIndizes, fullHarmonicBinList, harmPeaks, harmPeakFreqs):
    fig, ax = plt.subplots()
    arrowprops = dict(
        arrowstyle="->",
        connectionstyle="angle,angleA=0,angleB=90,rad=10")
    bbox = dict(boxstyle="round", fc="0.8")
    offset = 10

    plt.plot(faxis, 10 * log10(ps), c='r', label='noise')
    plt.plot(faxis[fundIndizes], 10 * log10(ps[fundIndizes]), c='b', label='$f_0$')
    for i, harm in enumerate(fullHarmonicBinList):
        plt.plot(faxis[harm], 10 * log10(ps[harm]), 'k')

        ax.annotate("f{}".format(i + 1),
                    (harmPeakFreqs[i], 10 * log10(harmPeaks[i])),
                    xytext=(0, 30), textcoords='offset points',
                    bbox=bbox, arrowprops=arrowprops)

    plt.legend()
    plt.show()

def snr_spectral(file):
    if os.path.isfile(file):
        sr, y = wavfile.read(file)
        if len(y.shape) > 1:  # Check if the audio is stereo and convert to mono if needed
            y = np.mean(y, axis=1)

        faxis, ps = sig.periodogram(y, fs=sr, window=('kaiser', 38))  # Get periodogram
        fundBin = argmax(ps)  # Estimate fundamental at maximum amplitude, get the bin number
        fundIndizes = getIndizesAroundPeak(ps, fundBin)  # Get bin numbers around fundamental peak
        fundFrequency = faxis[fundBin]  # Frequency of fundamental

        nHarmonics = 6
        harmonicFs = getHarmonics(fundFrequency, sr, nHarmonics=nHarmonics, aliased=True)  # Get harmonic frequencies
        print('Harmonic frequencies estimated: {}'.format(harmonicFs))

        # Search maxima near estimated harmonic frequencies
        harmonicBorders = np.zeros([2, nHarmonics], dtype=np.int16).T
        fullHarmonicBins = np.array([], dtype=np.int16)
        fullHarmonicBinList = []
        harmPeakFreqs = []
        harmPeaks = []

        for i, harmonic in enumerate(harmonicFs):
            searcharea = 0.1 * fundFrequency
            estimation = harmonic

            binNum, freq = getPeakInArea(ps, faxis, estimation, searcharea)
            harmPeakFreqs.append(freq)
            harmPeaks.append(ps[binNum])
            allBins = getIndizesAroundPeak(ps, binNum, searchWidth=1000)
            fullHarmonicBins = np.append(fullHarmonicBins, allBins)
            fullHarmonicBinList.append(allBins)
            harmonicBorders[i, :] = [allBins[0], allBins[-1]]
            print(freq)

        fundIndizes.sort()
        pFund = bandpower(ps[fundIndizes[0]:fundIndizes[-1]])  # Get power of fundamental
        fundRemoved = np.delete(ps, fundIndizes)  # Remove the fundamental (start constructing the noise-only signal)
        fAxisFundRemoved = np.delete(faxis, fundIndizes)

        plotSNR(ps, faxis, fundIndizes, fullHarmonicBinList, harmPeaks, harmPeakFreqs)

        noisePrepared = copy(ps)
        noisePrepared[fundIndizes] = 0
        noisePrepared[fullHarmonicBins] = 0
        noiseMean = median(noisePrepared[noisePrepared != 0])
        noisePrepared[fundIndizes] = noiseMean
        noisePrepared[fullHarmonicBins] = noiseMean

        noisePower = bandpower(noisePrepared)

        r = 10 * log10(pFund / noisePower)

        print('Measured SNR: {} dB'.format(r))
        print('Noise Power: {} dB'.format(10 * log10(noisePower)))

# Example
convert_mp3_to_wav(mp3_file_path, wav_file_path)
file_path = '/content/converted_audio.wav'
snr_spectral(file_path)
````
